```python
# Libraries and settings needed for analysis
import pandas as pd
import seaborn as sns

# Set figure size.
sns.set(rc={'figure.figsize':(15, 12)})

# Set the plot style as white.
sns.set_style('white')

# Maximum column width to display
pd.options.display.max_colwidth = 200
```

```python
# Load the data set.
tweets = pd.read_csv('tweets.csv')

# View the DataFrame.
print(tweets.shape)
print(tweets.columns)
tweets.head()
```

```python
# Explore the metadata.
print(tweets.dtypes)
```

```python
# Explore the data set.
trc = tweets['tweet_retweet_count'].value_counts(ascending=True)
trc
```

```python
tfc = tweets['tweet_favorite_count'].value_counts(ascending=True)
tfc
```

```python
# Would it be useful to only look at retweeted and favourite tweet messages?
# Explain your answer.
"""
It would not be a useful to look only at retweeted and favourite tweet messages because the main objective is to determine
whether there has been a high volume of hashtags related to healthcare in the UK. However, it might prove useful to see whether there is a relationship between these two types of messages and the count of hashtags of a certain word like "healthcare".
"""  
```

```python
# Create a new DataFrame containing only the text.
tweets_text = tweets[['tweet_full_text', 'tweet_entities_hashtags']]


# View the DataFrame.
tweets_text
```

```python
# Loop through the messages, and create a list of values containing the # symbol.
tags = []
for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z[z.index('#'):].lower())
```

```python
# Display the first 30 records.
hashtags = pd.Series(tags)
hashtags = hashtags.value_counts()
hashtags
```

```python
# Convert the series to a DataFrame in preparation for visualisation.
data = hashtags.to_frame()
data = data.reset_index()
data
```

```python
# Rename the columns.
data = data.rename(columns = {'index':'word', 0:'count'})
data
```

```python
# Fix the count datatype.
data.dtypes
```

```python
# Display records where the count is larger than 10.
records = data[data['count']>10]
records
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.
sns.barplot(x='count',y='word', estimator=sum, data=records)
```

```python
# Create the plot.
records = records.drop(index=0) # Drops the '#healthcare' datapoint as it is overrepresented 
records
```

```python
# View the barplot.
sns.barplot(x='count',y='word', estimator=sum, data=records)
```
